Interpretation:
A z-score of 1 means that the data point (75) is 1 standard deviation above the population mean (70). In a standard normal distribution, this corresponds to a position that is more than the mean but not extremely far from it.

In terms of probability, about 68.27% of values lie within one standard deviation of the mean in a normal distribution. Therefore, a score of 75 is not very unusual in this distribution. If we were looking at a standard normal distribution curve, this data point would be located to the right of the center, but not in the far right tail.
